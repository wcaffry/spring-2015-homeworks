```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import csv
import numpy as np
import sklearn.datasets as sk_data
import sklearn.metrics as metrics
from sklearn.cluster import KMeans
import scipy.sparse.linalg as linalg
import scipy.cluster.hierarchy as hr
from scipy.spatial.distance import pdist, squareform

```

```python
recipies = pd.read_csv("originalallr_recipes.txt", header= None, sep = '\n')
recipies.head()

recipies['Country'] = recipies[0].str.split('\t').str[0]
recipies['Ingredients'] = recipies[0].str.split('\t').str[1:]

"""First this reads the data file as a csv file and into a pandas DataFrame.  As the rows also include the country the next lines
split country and ingredients to separate columns"""

#print recipies.head()

recipies = recipies.drop(0, axis =1)
recipies= recipies.set_index(['Country'])
recipies['Ingredients']=recipies['Ingredients'].apply(lambda row: " ".join(row))

"""Drops the original column, sets the index to the country, and instead of listing the ingredients as a list in each row, converts
this list to a string for processing down the line"""

recipies.head()
```

```python
country_index=recipies.index.unique().tolist()

grouped_recipies=pd.DataFrame(index=country_index, columns=['Ingredients'])
grouped_recipies =grouped_recipies.reset_index()

"""Creates a new dataframe with each unique country or region as a row index and then converts them to their own column.
Ingredients are set as their own column"""

for idx, row in grouped_recipies.iterrows():
    row.Ingredients = recipies.ix[grouped_recipies['index'].ix[idx]]['Ingredients'].tolist()

grouped_recipies['Ingredients']=grouped_recipies['Ingredients'].apply(lambda row: " ".join(row))
print 

"""Takes all the recipes for each unique country and combines them into one list.  As above the list is converted into string
format so now the Ingredients column contains every recipe for that country in a long string for text parsing"""

grouped_recipies = grouped_recipies[grouped_recipies['index'] != 'American']
chart_index=grouped_recipies['index'].tolist()
#len(chart_index)

"""Drops all American recipies as there are an order of magnitude more of those than any other and as the majority of the country 
were or were decended from imigrants at somepoint may throw off grouping.  Also creates a list of the Countries in the dataframe"""

grouped_recipies.head()
#print grouped_recipies[grouped_recipies['index']=='Canada']
```

```python
#recipies.groupby(level=0).size()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df=1, max_df=.99)
recp_vec = vectorizer.fit_transform(grouped_recipies['Ingredients'])

"""Runs Tfidf vectorizer on the Ingredients column in the grouped_recipies dataframe creating a matrix with countries as rows and 
ingredients and there frequencies in the columns"""

print type(recp_vec), recp_vec.shape
#print recp_vec
```

In the vectorizer the min_df is set as low as possible to pick up on ingredients that may be specific to a certain culture and not seen in other cultures, which therefore have value in differentiating that culture.  The max_df is set very high as some ingredients will be seen in almost every culture (like eggs), but the frequency of which they occur in recipies of that culture may contain informative characteristics. 

```python
distanceMatrix = pdist(recp_vec.todense(),'euclidean')
Z = hr.linkage(distanceMatrix,  method='complete', metric='euclidean')

#print distanceMatrix
```

```python
fig = plt.figure(figsize=(15,15))
T = hr.dendrogram(Z,color_threshold=0.9, labels =chart_index, leaf_font_size=10)

"""Dendrogram showing relationship of countries based on their recipies"""

fig.show()
```

```python
ck = hr.fcluster(hr.linkage(distanceMatrix, method='complete'), 1.36,'distance')
print ck
```

```python
distances = metrics.euclidean_distances(recp_vec)
cg = sns.clustermap(distances, method="complete", figsize=(0.1,0.1))

heatmap_index=[]
for i in range(len(cg.dendrogram_col.reordered_ind)):
    heatmap_index.append(chart_index[cg.dendrogram_col.reordered_ind[i]])

"""Creates a list of the labels for the dendrogram by pulling their index from the dendrogram and calling the appropriate index
from the list of countries so the chart is labeled correctly"""
    
print heatmap_index
cg = sns.clustermap(distances, method="complete", figsize=(13,13), yticklabels=heatmap_index)
```

As can be seen in the two trees above the northern and eastern Eropean regions cluster together in both trees.  Evident from the heatmap above is the lighter color of the square representing these 10 countries, indicating not only do the eastern and northern European countries cluster together, but that their cuisine is more similar to each other than cuisines in the other large cluster block (mainly asian and african) are to other cuisines within the afro-asian cluster.  

```python
from collections import Counter

canada = grouped_recipies[grouped_recipies['index']=='Canada']['Ingredients'].values.tolist()
canada = canada[0].split()
print 'Canadas top 10 ingredients', Counter(canada).most_common(10)

uk_and_ireland = grouped_recipies[grouped_recipies['index']=='UK-and-Ireland']['Ingredients'].values.tolist()
uk_and_ireland = uk_and_ireland[0].split()
print 'UK and Ireland\'s top 10 ingredients', Counter(uk_and_ireland).most_common(10)

switzerland = grouped_recipies[grouped_recipies['index']=='Switzerland']['Ingredients'].values.tolist()
switzerland = switzerland[0].split()
print 'Switzerland top 10 ingredients', Counter(switzerland).most_common(10)

italy = grouped_recipies[grouped_recipies['index']=='Italy']['Ingredients'].values.tolist()
italy = italy[0].split()
print 'Italy top 10 ingredients', Counter(italy).most_common(10)

philippines = grouped_recipies[grouped_recipies['index']=='Philippines']['Ingredients'].values.tolist()
philippines = philippines[0].split()
print 'Philippines top 10 ingredients', Counter(philippines).most_common(10)
```

Looking at Italy and the Philippines we can see that while they share some ingredients the order in which they appear is mostly different and there are more unique ingredients between the two.  So even though they may have been clustered in the same large cluster, they exhibit a fair amount of difference between one another.

In contrast, looking at the top ingredients for Canada, UK and Switzerland, there are a large number of shared ingredients that show up and in relatively the same order.  This makes sense given the heatmap above showing this group exhibits strong similarities to one another.  So what makes the anglo/nordic/germanic food anglo/nordic/germanic?  Judging by the top ten ingredients for Canada, UK, and Switzerland, a preponderance of wheat, butter, egg, and milk.  And with those ingredients one can only think, pancakes anyone?  Add a glob of maple syrup and what could be more Canadian?
