## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
top_occurence = high_occurence.sort('occurences', ascending = False).iloc[0:5]
top_occurence
````

````python
# Write your code here
import numpy as np
case_by_day =pd.DataFrame(index=np.arange(0, len(boston.groupby(boston.index.map(lambda datetime: datetime.date())))), 
                          columns =['Request for Snow Plowing', 'Snow Removal', 
                                    'Requests for Street Cleaning', 'Request for Pothole Repair', 'Misc. Snow Complaint',
                                    'Day'])
"""Create empty data frame with 5 columns and index set to length of the boston dataframe grouped by day"""

def days_summary(data_frame):
    index=0
    for day, day_group in data_frame.groupby(data_frame.index.map(lambda datetime: datetime.date())):
        """Iterates through each day in dataframe"""
        
        days_cases = day_group.groupby('case_title')
        days_cases = pd.DataFrame(days_cases.size())
        days_cases = days_cases.reset_index()
        days_cases.columns = ['case_title', 'count']    
        """Creates new dataframe for each day which contains each case title and its number of appearances in the days dataframe"""
        
        for idx, row in days_cases.iterrows():
            if row.case_title =='Snow Removal':
                case_by_day.loc[index]['Snow Removal'] = row['count']
            elif row.case_title == 'Request for Snow Plowing':
                case_by_day.loc[index]['Request for Snow Plowing'] = row['count']
            elif row.case_title == 'Requests for Street Cleaning':
                case_by_day.loc[index]['Requests for Street Cleaning'] = row['count']
            elif row.case_title == 'Request for Pothole Repair':
                case_by_day.loc[index]['Request for Pothole Repair'] = row['count']
            elif row.case_title == 'Misc. Snow Complaint':
                case_by_day.loc[index]['Misc. Snow Complaint'] = row['count']
        case_by_day.loc[index]['Day'] = day
        index +=1
        """Iterates through the days dataframe and searches for keywords, then appends their value to the keyword column in a new
        dataframe at the days index"""
        
    case_by_day.fillna(0, inplace = True)
    case_by_day.index=case_by_day['Day']
    return case_by_day
    """replaces NaN with a zero count as NaN means that call did not occur on that day, and resets the index to the 'Day' column""" 
    
days_summary(boston)
case_by_day = case_by_day.drop(['Day'], axis =1)


fig, axes = plt.subplots()
case_by_day.plot(ax=axes)


axes.set_title("Call Requests by Day")
axes.set_ylabel("# Calls")
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
snow_data.head()
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# Write your code here
snow_data.index = snow_data['DATE']
noerror_sdata=snow_data[snow_data['SNOW']>=0]
"""creates new dataframe filtering out instrument errors that result in negative precipitation readings"""

agg_mean=noerror_sdata.groupby(noerror_sdata.index.map(lambda datetime: datetime.date())).mean()
agg_mean['SNOW']=agg_mean['SNOW'].apply(lambda x: x/10)
"""creates new dataframe that first groups the noerror_sdata dataframe by date, then takes the mean snow reading from all the 
stations in boston for that day and converts it from millimeters to centimeters"""

axes =plt.subplot()
agg_mean.ix[:,1].plot(ax = axes, color ='gray', legend=True)
"""plots only the 'SNOW' column against the date"""

axes.set_ylabel('Snow in CM')
axes.set_title('Snowfall by Day')
axes.set_xlabel('Date')
#agg_mean
````

## Step 4. Matching trends

The last thing we will do is to join the snowfall dataframe with that of the top 5 categories of calls, per day. Don't forget to fill the NaN values with 0. You want to do an outer join (if there are calls for a day but no snowfall data, we want to create a 0 record). Save that to a dataframe called `calls_on_snow`. The first 5 columns should be the counts for the categories, and the 6th will be the snowfall data. **(10 pts)**

````python
calls_on_snow=case_by_day
calls_on_snow=calls_on_snow.join(agg_mean['SNOW'])
"""apends the agg_mean 'SNOW' column to the case_by_day dataframe"""
calls_on_snow.head()
````

Now, you can show the connection of the two trends (the calls and the snow) in a plot. I provide some sample code, but you should also try your own versions.

````python
fig, axes = plt.subplots(nrows=2,figsize=(12,5), sharex=True)

calls_on_snow.ix[:,0:5].plot(ax=axes[0])
calls_on_snow.ix[:,5].plot(ax=axes[1],label="snowfall (in cm)", color='gray')

axes[0].set_title("The effect of the winter snowstorms on Boston's 311 Hotline for 2015")
axes[0].set_ylabel("# Calls")
axes[1].set_ylabel("Snowfall in cm")
axes[1].set_xlabel("Date")
axes[0].xaxis.grid(False)
axes[1].xaxis.grid(False)
````
