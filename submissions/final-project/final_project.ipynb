```python
def distance(lat1, long1, lat2, long2):
    R = 6371 # Earth Radius in Km
    dLat = math.radians(lat2 - lat1) # Convert Degrees 2 Radians
    dLong = math.radians(long2 - long1)
    lat1 = math.radians(lat1)
    lat2 = math.radians(lat2)
    a = math.sin(dLat/2) * math.sin(dLat/2) + math.sin(dLong/2) * math.sin(dLong/2) * math.cos(lat1) * math.cos(lat2)
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1-a))
    d = R * c
    return d 

print distance(eagle.iloc[0]['lat'], eagle.iloc[0]['long'], eagle.iloc[1]['lat'], eagle.iloc[1]['long'])

print distance(gegret.iloc[0]['lat'], gegret.iloc[0]['long'], gegret.iloc[1]['lat'], gegret.iloc[1]['long'])

print distance(bluheron.iloc[0]['lat'], bluheron.iloc[0]['long'], bluheron.iloc[1]['lat'], bluheron.iloc[1]['long'])
```

```python
def get_distance(df):
    dist = [0]
    for i in range(0, len(df)-1):
        x = distance(df.iloc[i]['lat'], df.iloc[i]['long'], df.iloc[i+1]['lat'], df.iloc[i+1]['long'])
        dist.append(x)
    return dist

eagledist = get_distance(eagle)
herondist = get_distance(bluheron)
gretdist = get_distance(gegret)
```
